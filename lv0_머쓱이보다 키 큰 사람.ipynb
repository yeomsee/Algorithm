Q. 머쓱이는 학교에서 키 순으로 줄을 설 때 몇 번째로 서야하는지 궁금해졌습니다.  

머쓱이네 반 친구들의 키가 담긴 정수 배열 array와 머쓱이의 키 height가 매개변수로 주어질 때, 머쓱이보다 키 큰 사람 수를 return 하도록 solution 함수를 완성해보세요.

# 나의 코드

```python
def solution(array, height):
    answer = len([x for x in array if x > height])
    return answer
```

---

# 비슷한 코드

```python
def solution(array, height):
    return sum(1 for a in array if a > height)
```

# BEST 코드

시간복잡도는 비슷하지만 array만 쓰라고 하면 이게 best일 듯

```python
def solution(array, height):
    array.append(height)
    array.sort(reverse=True)

    return array.index(height)
```
